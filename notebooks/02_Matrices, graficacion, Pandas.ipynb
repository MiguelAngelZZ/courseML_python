## 2. Linspace, matrices y graficación

El comando **linspace** crea una matriz lineal de puntos desde un valor inicial hasta un valor final.

```python
import numpy as np
np.linspace(0,2)
```

```python
np.linspace(0,2,10)
```

**linspace** es una forma sencilla de crear coordenadas para trazar gráficas ya que las funciones de la biblioteca numpy  pueden actuar sobre un vector completo (o una matriz) de puntos a la vez. Así,

```python
x = np.linspace(0, 5, 50)
np.sin(x)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
plt.plot(x,np.sin(x))
```

## Montecarlo y Números aleatorios 

Muchos métodos de computación científica se basan en la integración de Monte Carlo, donde se utiliza una secuencia de números pseudo-aleatorios para aproximar la integral de una función. Python tiene excelentes generadores de números aleatorios en la biblioteca estándar. La función random() genera números pseudoaleatorios distribuidos uniformemente entre 0 y 1:

```python
from random import random
alea = []
for i in range(100):
    alea.append(random())
plt.plot(alea, 'bo')
```

```python
from random import gauss
g_alea = []
for i in range(100):
    g_alea.append(gauss(0,1))
plt.plot(g_alea, '-r')
```

```python
plt.plot(np.random.rand(100))
```

### Series

Una serie es un objeto similar a una matriz unidimensional que contiene una matriz de datos y una matriz asociada de etiquetas de datos. Los datos pueden ser de cualquier tipo de datos NumPy y las etiquetas son el índice de la serie.

```python
seri1 = pd.Series([2, 0, 1, 3, -5, 18, 7])
seri1
```

```python
seri1.values ## representación en array
```

```python
seri1.index ### indice de la serie
```

```python
seri2 = pd.Series([2, 0, 1, 3, -5, 18, 7], index=['a', 'b', 'c', 'd', 'e', 'f', 'g'])
seri2
```

```python
seri2[2] == seri2['c'] ## obteniendo valores de la serie
```

```python
seri2[['a', 'g', 'b']] ## indexar valores pasando una lista
```

```python
seri2[seri2 > 1] ### filtrar valores mayores a cero
```

```python
seri2 * 5 ## multiplicacion escalar
```

```python
np.exp(seri2) ### aplicadno una función Numpy
```

```python
pd.isnull(seri2).sum() ## detectando valores perdidos
```

## Dataframe
Un DataFrame es un arreglo de datos rectangular que contiene una colección ordenada de columnas (panda series). Cada columna puede ser de un tipo diferente. Tiene índices de fila y columna. Las operaciones de filas y columnas se tratan aproximadamente de forma simétrica. Las columnas devueltas al indexar un DataFrame son vistas de los datos subyacentes y no una copia. Para obtener una copia, se debe usar el método de copia de la serie.

```python
data1 = {'mpio' : ['Bogotá', 'Medellín', 'Cali', 'Barranquilla', 'Bucaramanga'],
          'año' : [2012, 2013, 2014, 2014, 2015],
          'indice' : [15.1, 18.3, 12.5, 9.7, 5.3]}
df1 = pd.DataFrame(data1)
df1
```

```python
df_2 = pd.DataFrame(data1, columns=['año', 'mpio', 'indice'])
df_2
```

```python
df_3 = pd.DataFrame(data1, columns=['año', 'mpio', 'indice', 'poblacion']) ### sise añade una columna que noexiste devuelve NA
df_3
```
